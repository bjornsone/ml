### Activation Functions
Includes: sigmoid, tanh, elu, relu, softplus, and softsign <br/>
[Tensorflow Documentation](https://www.tensorflow.org/versions/r0.12/api_docs/python/nn/activation_functions_)

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Define the range over which we are going to graph
xs_np = np.arange(-2, 2, .1)
xs_tf = tf.Variable(xs_np)

activation_functions= {}
activation_functions['sigmoid'] = tf.sigmoid(xs_tf)
activation_functions['relu'   ] = tf.nn.relu(xs_tf)
activation_functions['tanh'   ] = tf.tanh   (xs_tf)
activation_functions['elu'    ] = tf.nn.elu (xs_tf)


with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    results = {}
    for name, node_tf in activation_functions.items():
        results[name] = sess.run(node_tf)

plt.figure(figsize=(5, 5))
for _, ys_np in results.items():
    plt.plot(xs_np, ys_np, '-');
plt.legend(results.keys(), loc='upper left');
```
